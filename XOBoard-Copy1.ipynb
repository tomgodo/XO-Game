```python
############################################
##########      X O BOARD       ############
############################################


# board is a dictionary 
# 9 keys 1-9 (int)
# all values are ' '

class XOBoard:

    # this function will be used to
    # return the status of the board
    # 4 possibilities: 
    # 'X' if x wins
    # 'O' if o wins
    # 'None' if the game is not finished
    # 'Draw' if there is no winner
    
    def status (board):
        
        # patterns of winning possibilities
        patterns = [[1,2,3],[4,5,6],[7,8,9],[1,4,7],[2,5,8],
                    [3,6,9],[1,5,9],[3,5,7]]
        
        # create 2 lists of 'O' & 'X' positions on the board 
        O_list = []
        X_list = []
        for i in board.keys():
            if board[i] == 'O':
                O_list.append(i)
            if board[i] == 'X':
                X_list.append(i)
        ##print (O_list)
        ##print (X_list)
        
        # check for 'O' win & for 'X' win
        # comparing between every win pattern
        # and current positions lists ('X' & 'O')
        
        i = 0
        # patterns loop (go through all patterns - 9)
        while i < 8:
            # check if there is a win possibility only (5 moves at least)
            if len(X_list) + len(O_list) < 5:
                break
            # pattern initialization    
            pattern = patterns[i]
            # indexes initialization 
            i_O = 0 # positions list index
            j_O = 0 # pattern index
            # pattern loop (go through the 3 pattern posiyions)
            while j_O < 3:
                # icrease both indexes if found a match
                if O_list[i_O] == pattern[j_O]:
                    i_O += 1
                    j_O += 1
                # increase positions list only
                else:
                    i_O += 1
                # end comparision if riched the end of poaitions list
                if i_O == len(O_list):
                    break
            # retur result if reach 3  matches
            if j_O == 3:
                return 'O'
            
            i_X = 0
            j_X = 0
            while j_X < 3:
                if X_list[i_X] == pattern[j_X]:
                    i_X += 1
                    j_X += 1
                else:
                    i_X += 1
                if i_X == len(X_list):
                    break
            if j_X == 3:
                return 'X'
                    
            i +=1
        # check for free positions on the board
        for k in board.keys():
            # return 'None' (game not finished) if fond free position            
            if board[k] == ' ': 
                return None
        # return 'Draw' if the board is full
        return 'Draw'
       
        
    # this function will be used to
    # put a simbol on the board
    # the function recive 3 arguments:
    # 1 - the board
    # 2 - the simbol ('X' or 'O' only)
    # 3 - the position (1 - 9)
    
    def put (board, myinput, position):
        
        try:
            # transform the simbol input in capital letter
            xo = myinput.upper()
            # check if the simbol is legal
            if xo == 'X' or xo == 'O':
                #check if the position is free
                if board[position] == ' ':
                    #print (board[position-1])
                    # put the mouve on the board
                    board[position] = myinput.upper()
                else:
                    print ('Invalid mouve, pleas chouse e free place')
            else:
                print ("Invalid simbol, use 'X' or 'O' only")
            return
        except:
            print ('Positions: 1-9 only')
    
    # this function will be used to
    # print the board on the screen
    # include players moves
    def display (board):
        ##print (board[1], board[2], board[3],board[4], board[5], board[6], board[7], board[8], board[9])
        print ("   |   |   \n {0:^1} | {1:^1} | {2:^1} \n   |   |   \n-----------\n   |   |   \n {3:^1} | {4:^1} | {5:^1} \n   |   |   \n-----------\n   |   |   \n {6:^1} | {7:^1} | {8:^1} \n   |   |   \n" .format 
               (board[1], board[2], board[3],board[4], board[5], board[6], board[7], board[8], board[9]))
        return
    

    
# execute code
'''board = {1:' ', 2:' ', 3:' ', 4:' ', 5:' ', 6:' ', 7:' ', 8:' ', 9:' '}

myinput = input ("Enter 'x' or 'o'")
position = input ("Enter a number 1-9")
XOBoard.put (board, myinput, int(position))
XOBoard.display (board)
status = XOBoard.status (board)
print (status)'''
```
